## Data visualization

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)

train_df = pd.read_csv('traininingdata.txt', sep=";")
test_df = pd.read_csv('testdata.txt', sep=";")
```

```python
numeric_var = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
binary_var = ['y', 'default', 'housing', 'loan']
categorical_var = ['job', 'marital', 'education', 'contact', 'month', 'poutcome']
```

```python
# binary feature visualization
def visual_binary(var):
    fig, ax = plt.subplots(1, 2, figsize=(16, 8))
    sns.histplot(
            x=var,
            alpha=0.8,
            hue=train_df.columns[-1], 
            data=train_df,
            ax=ax[0],
            multiple='stack',
        ).set_title('train_df')

    sns.histplot(
            x=var,
            alpha=0.8,
            hue=test_df.columns[-1], 
            data=test_df,
            ax=ax[1],
            multiple='stack',
        ).set_title('test_df')

    # save the plot
    plt.savefig(f'./images/{var}.png')
```

```python
for var in binary_var:
    visual_binary(var)
```

```python
# numeric feature visualization
def visual_numeric(var):
    fig, ax = plt.subplots(1, 2, figsize=(16, 8))
    sns.histplot(
            x=var,
            alpha=0.8,
            hue=train_df.columns[-1], 
            data=train_df,
            ax=ax[0],
            # stat="probability",
            multiple='stack',
        ).set_title('train_df')

    sns.histplot(
            x=var,
            alpha=0.8,
            hue=test_df.columns[-1], 
            data=test_df,
            ax=ax[1],
            # stat="probability",
            multiple='stack',
        ).set_title('test_df')
    # save the plot
    plt.savefig(f'./images/{var}.png')
```

```python
## numeric_var = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
for var in numeric_var:
    visual_numeric(var)
```

```python
def visual_cate(var):
    fig, ax = plt.subplots(1, 2, figsize=(16, 8))
    
    if var == 'job':
        ax[0].tick_params(labelrotation=25)
        ax[1].tick_params(labelrotation=25)
        
    sns.histplot(
            x=var,
            hue=train_df.columns[-1], 
            multiple="dodge",
            shrink=0.7,
            data=train_df,
            ax=ax[0]
        ).set_title('train_df')
    
    
    sns.histplot(
            x=var,
            hue=test_df.columns[-1], 
            multiple="dodge",
            shrink=0.7,
            data=test_df,
            ax=ax[1]
        ).set_title('test_df')

    # save the plot
    plt.savefig(f'./images/{var}.png')
```

```python
## categorical_var = ['job', 'marital', 'education', 'contact', 'month', 'poutcome']
for var in categorical_var:
    visual_cate(var)
```
